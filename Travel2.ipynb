# More Travel Maps

Country shapefiles from gadm.org

```python
import travelmaps as tm
from adashof import cm2in
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

## Global Map

```python
fig_x = plt.figure(figsize=(cm2in([12, 8])))

# Create basemap
m_x = Basemap(resolution='i', projection='robin', lon_0=0)
m_x.drawlsmask(land_color='.8', ocean_color='#f7fbff')
m_x.drawcoastlines(color='.8', linewidth=.1)
m_x.drawcountries(color='.8', linewidth=.1)

# Fill visited countries
countries = ['CHN', 'CHE', 'DNK', 'TUR', 'MEX', 'HUN', 'ISL', 'BGR', 'GRC', 'FRA',
             'SVK', 'NOR', 'CZE', 'AUT', 'DEU', 'GBR', 'MAR', 'SRB', 'USA', 'RUS',
             'ESP', 'IRL', 'AUS', 'BRA', 'EGY', 'SWE', 'ITA', 'BEL', 'NLD', 'SYR']
tm.country(countries, m_x)

#plt.savefig('plots/World2.png', bbox_inches='tight', pad_inches=0.02)

plt.show()
```
