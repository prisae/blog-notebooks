## gridfill()




```python
import numpy as npx
import matplotlib as mpl
import matplotlib.pyplot as plt
# Increase font size,
mpl.rc('font', size=16)
# Define colours (taken from http://colorbrewer2.org)
clr = ['#377eb8', '#e41a1c', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628']
plt.show()
```

### Load the `gridfill`-function

```python
from matplotlib import colors

def gridfill(xval, yval, values, style='colour', cmap=mpl.cm.Spectral, unicol='#000000', lc='k', lw=0.5):
    """Fill rectangular grid with colours or a colour and transparency.

    Parameters
    ----------
    xval, yval : array
        Grid-points in x- and in y-direction.
    values : array, dimension: x*y
        Values between 0 and 1
    style : string, optional; {<'colour'>, 'alpha'}
        Defines if values represent colour or alpha.
    lc, lw : 
        Line colour and width, as in standard plots.

    Returns
    -------
    locs : list
        List of plotted polygon patches.
    """
         
    def parr(one, two, case):
        """Create 2D arrays for fill-polygons."""
        if case == 0:
            onerep = np.array([one[:-1], one[:-1], one[1:], one[1:]])
        elif case == 1:
            onerep = np.array([one[:-1], one[1:], one[1:], one[:-1]])
        
        return onerep.repeat(len(two)-1,axis=case).reshape(4,-1)


    if style == 'alpha':
        unicol = colors.colorConverter.to_rgb(unicol)
        colour = np.vstack((np.repeat(unicol, len(values)).reshape(3,-1), values)).transpose()
    else:
        
        valnan = np.isnan(values)
        values[valnan] = 0
        cmcol = cmap(np.linspace(0,1,101))
        colour = cmcol[list(map(int, 100*values))]
        colour[valnan, -1] = 0 # set alpha to 0
        
    rct = plt.gca().fill(parr(xval, yval, 0), parr(yval, xval, 1), lw=lw, ec=lc)
    cls = list(map(colors.rgb2hex, colour))
    
    for ind in range(len(rct)):
        rct[ind].set_facecolor(cls[ind])
        rct[ind].set_alpha(colour[ind, -1])
        
    return rct

   
```

```python
# Create some data
# X-direction 20, Y-direction 10, Z-direction 3
# X/Y-direction random spacing {1, 2, 3}
x = np.cumsum(np.r_[0, np.random.randint(1,4,19)])
y = np.cumsum(np.r_[0, np.random.randint(1,4,9)])
z = np.arange(3)

maxval = 10
val = np.random.rand(len(x), len(y), len(z))*(maxval+5)-5
val[val<=0] = np.NaN


# Define colours for resistivity plots.
cmap = mpl.cm.Spectral
cmcol = cmap(np.linspace(0,1,101))
cmcol[0,-1] = 0 # Set alpha of first entry to 0


def create_figure(y=y):
    fig = plt.figure(figsize=(8, 5))
    ax = plt.gca()
    ax.set_axis_off()
    ax.axis('equal')
    ax.set_xlim([-.5, max(x)+.5])
    return fig, ax

    
    
# Make a figure and axes with dimensions as desired.
fig0 = plt.figure(figsize=(8, .3))
ax0 = fig0.add_axes([0.05, 0.05, 0.9, 0.9])
cb1 = mpl.colorbar.ColorbarBase(ax0, cmap=cmap, orientation='horizontal',
                                norm=mpl.cm.colors.Normalize(vmin=0, vmax=10))
                                
```

### 1. Top slice with black lines

```python
fig1, ax1 = create_figure()
inpval = val[:,:,0].ravel()/maxval
rct = gridfill(x, y, inpval, 'colour')
```

### 2. Middle slice with white lines

```python
# 2. Middle slice
fig2, ax2 = create_figure()
inpval = val[:,:,1].ravel()/maxval
rct = gridfill(x, y, inpval, 'colour', lc='w', lw=1)
```

### 3. Bottom slice, without lines

Create `gridfill` without lines of the bottom slice.

```python
fig3, ax3 = create_figure()
inpval = val[:,:,2].ravel()/maxval
rct = gridfill(x, y, inpval, 'colour', lc='none')
```

### 4. Stacked, transparency reflects number of cells

```python
fig4, ax4 = create_figure()

# Calculate the accumulated cell-density
# 1. Create a copy of val
alphaval = np.ones(np.shape(val))
# 2. 1 where val>0 , else 0
alphaval[np.isnan(val)] = 0
# 3. Sum and normalize
alphaval = alphaval.sum(axis=2).ravel()/len(z)

# Plot the result
rct = gridfill(x, y, alphaval, style='alpha', unicol=clr[0], lc=clr[0], lw=2)

```
