## move_sn_y()

The following function, `move_sn_y(offs=0, dig=0, side='left', omit_last=False)`, moves the scientific exponent from the top to the side. Possible reasons why you want to move your exponent from the top to the side:

- The exponent overlaps with your figure title.
- Your figure has no title; instead, you include the figure in a document, e.g. $\LaTeX$, and create the title figure there. Having the scientific notation on top of the figure creates then unnecessary white space between your title and the plot.
- You simply want your figure with scientific notation to appear the exactly same as a figure without it.

For me personally, it is a combination of the latter two. Now, a word of caution: this function is not terrible stable and fool-proof. Specifically, it does not thoroughly check the input and all possible cases; expect therefore failures if you don't treat it as it expects it.

```python
import urllib3
import numpy as np
from matplotlib import rc
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from adashof import circle

# Increase font size, set CM as default text, and use LaTeX
rc('font', **{'size': 16, 'family': 'serif', 'serif': ['Computer Modern Roman']})
rc('text', usetex=True)

# Define colours (taken from http://colorbrewer2.org)
colours = ['#377eb8', '#e41a1c', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628']
```

```python
def move_sn_y(offs=0, dig=0, side='left', omit_last=False):
    """Move scientific notation exponent from top to the side.

    Parameters
    ----------
    offs : float, optional; <0>
        Horizontal movement additional to default.
    dig : int, optional; <0>
        Number of decimals after the comma.
    side : string, optional; {<'left'>, 'right'}
        To choose the side of the y-axis notation.
    omit_last : bool, optional; <False>
        If True, the top y-axis-label is omitted.

    Returns
    -------
    locs : list
        List of y-tick locations.

    Note
    ----
    This is kind of a non-satisfying hack, which should be handled more
    properly. But it works. Functions to look at for a better implementation:
    ax.ticklabel_format
    ax.yaxis.major.formatter.set_offset_string
    """

    locs,_ = plt.yticks()

    # If locs are not in scientific notation, put them into it to find exponent
    try:
        yoff = int(str(locs[-1]).split('e')[1])
    except:
        llocs = '%.3e' % locs[-1]
        yoff = int(str(llocs).split('e')[1])
    
    form = r'$%.'+str(dig)+'f$'
    if omit_last:
        slocs = locs[:-1]
    else:
        slocs = locs

    plt.yticks(locs, list(map(lambda x: form % x, slocs/(10**yoff))))

    if side == 'left':
        offs = -.18 - offs
    elif side == 'right':
        offs = 1 + offs
    plt.text(offs, 1, r'$\times10^{%i}$' % yoff, transform =
            plt.gca().transAxes, verticalalignment='top')

    return locs
```

### Examples showing how to use it
For my example I played a bit with web-crawling. The actual plots are not necessarily meaningful, their solely purpose is to show the use of `move_sn_y`.

```python
# Get some data to plot:
# A list of countries with corresponding population and area
datasite = 'http://www.worldatlas.com/aatlas/populations/ctyareal.htm'
# Worldatlas states the following data sources:
# CIA World Factbook, and other public domain resources (February 2006)

# Read the page
http = urllib3.PoolManager()

# BeautifulSoup is the beast that makes the html bette accessible to parse
soup = BeautifulSoup(http.request('GET', datasite).data)

# We are interested in the second table
# (Have a look at the entire source code of the webpage to find out
# which table you need -> print(soup))
table = soup('table')[1]
```

Now we can parse this table to get the data we are interested in:

```python
# Pre-allocate arrays and list
population = np.zeros(192)
area = np.zeros(192)
country = [0]*192

# Loop through rows and cells to get thi information
i = 0
for row in table.findAll("tr"):
    cells = row.findAll("td")
    # The is a hidden column in which we are not interested, and title columns
    # have no text in the first cell, so we exclude them too.
    if cells[0].find(text=True) != None and row['style'] != 'display:none':
        country[i] = cells[1].find(text=True)
        population[i] = np.array(cells[2].find(text=True).replace(',', ''), dtype=float)
        area[i] = np.array(cells[3].find(text=True).replace(',', ''), dtype=float)
        i += 1
```

First a QC plot to check the data.

```python
# Create figure
fig1 = plt.figure()
  
# Plot data on a loglog-scale
plt.loglog(population, area, 'o', c=colours[0], mec='.8', ms=8)

# Highlight some countries
i = 0
for count in ['Mexico', 'United Kingdom (UK)', 'Switzerland']:
    ci = country.index(count)
    plt.loglog(population[ci], area[ci], '*', c=colours[i+1], mec='k', ms=14, label=count)
    i += 1
    
# Label the plot and set limits
plt.axis([2e2, 8e9, 0, 8e7])
plt.title('Countries plotted as population versus area')
plt.xlabel(r'Population')
plt.ylabel(r'Area ($\rm{km}^2$)')
plt.legend(loc='lower right', frameon=False, fontsize=12, numpoints=1)

plt.show()
```

I now plot the area as

```python
# small amplitudes (power of minus ten)

fig2a = plt.figure()
  
plt.plot(population[-18:]/100000, area[-18:], 'o', c=colours[0], mec='.8', ms=8)
#/sum(area)
i = 0
for count in ['San Marino', 'Malta', 'Andorra']:
    ci = country.index(count)
    plt.plot(population[ci]/100000, area[ci], '*', c=colours[i+1], mec='k', ms=14, label=count)
    i += 1
    
plt.title(r'18 smallest countries as fraction of total land masses')
plt.xlabel(r"Population (in 100'000s)")
plt.ylabel(r'Area fraction')
plt.legend(loc='lower right', frameon=False, fontsize=14, numpoints=1)

# Set scientific notation
plt.gca().yaxis.major.formatter.set_powerlimits((0,0)) 

plt.axis([-.2, 4.3, -50, 750])
plt.draw()
```

```python
fig2b = plt.figure()
  
plt.plot(population[-18:]/100000, area[-18:], 'o', c=colours[0], mec='.8', ms=8)

i = 0
for count in ['San Marino', 'Malta', 'Andorra']:
    ci = country.index(count)
    plt.plot(population[ci]/100000, area[ci], '*', c=colours[i+1], mec='k', ms=14, label=count)
    i += 1
    
plt.title(r'18 smallest countries as fraction of total land masses')
plt.xlabel(r"Population (in 100'000s)")
plt.ylabel(r'Area fraction')
plt.legend(loc='lower right', frameon=False, fontsize=14, numpoints=1)

# Set scientific notation
move_sn_y(offs=-.03, side='left')

plt.axis([-.2, 4.3, -50, 750])
plt.draw()
```
