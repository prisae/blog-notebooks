## Labels etc

```python
# Nr of months and Januarys, months as number-array, decreasing
nrm = np.size(data[:,0])
nrjans = int((nrm-4)/12+1)
months = np.arange(nrm, 0, -1)-1

# Label: empty string except for Januarys
mts, yrs = np.modf(data[:,0])
janlab = []
for i in months[4::12]:
    janlab += [datetime.date(int(yrs[i]), int(round(100*mts[i])), 1).strftime('%b \'%y')]
label = nrm*['']
label[4::12] = janlab[:nrjans]
label = label[-1::-1] # (reverse, like months and data)

# Labelticks
labeltic = np.array(label)
labeltic[labeltic != ''] = months[labeltic != '']
labeltic[labeltic == ''] = '0'
labeltic = labeltic.astype(float)

# Get start and end month name and year for plot titles
smtsyr = datetime.date(int(yrs[-1]), int(round(100*mts[-1])), 1).strftime('%b \'%y')
emteyr = datetime.date(int(yrs[0]), int(round(100*mts[0])), 1).strftime('%b \'%y')
```

## Plots

```python
fig1, ax = plt.subplots()

# Plot the data
plt.plot(months, data[:,1]/data[-1,1]*100, label='CHF')
plt.plot(months, data[:,2]/data[-1,2]*100, label='GBP')
plt.plot(months, data[:,3]/data[-1,3]*100, label='USD')
plt.plot(months, data[:,4]/data[-1,4]*100, label='EUR')
plt.plot(months, data[:,5]/data[-1,5]*100, label='BRL')

# Title, labels, legend, axis limits
plt.title(('Value of the Mexican Peso ('+smtsyr+' - '+emteyr+')'))
plt.ylabel('Percent %')
plt.legend(loc=3, ncol=2)
ax.axis([0, np.size(months)-1, 80*np.min(data[:,1:]/data[-1,1:]),  105*np.max(data[:,1:]/data[-1,1:])])
ax.set_xticks(labeltic)
ax.set_xticklabels(label)

# Save and show
plt.savefig('plots/MXNper.png', bbox_inches='tight')
plt.show()
```

```python
fig2, ax1 = plt.subplots()

# axis limits
axis = [0, np.size(months)-1, .9*np.min(data[:,1]),  1.05*np.max(data[:,1])]

# Plot data and set axis limits
ax1.plot(months, data[:,1], 'k')
ax1.axis(axis)

# Set label blue (1 MXN - x CHF)
for tl in ax1.get_yticklabels():
    tl.set_color('b')
ax1.set_ylabel('1 MXN = x CHF', color='b')

# Second plot for second label, without grid, same limits
ax2 = ax1.twinx()
ax2.grid('off')
ax2.axis(axis)

# Set label red (1 CHF - x MXN)
ax2.set_ylabel('1 CHF = x MXN', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')

# Get tick labels, and invert them
# NOTE: inverting them makes this scale non-linear!
labels = np.round(1/np.array(ax2.get_yticks().tolist()),1)
ax2.set_yticklabels(labels)

# Title and x-axis ticks
plt.title(('MXN/CHF exchange rates ('+smtsyr+' - '+emteyr+')'))
ax2.set_xticks(labeltic)
ax2.set_xticklabels(label)    

# Save and show
plt.savefig('plots/MXNexr.png', bbox_inches='tight')
plt.show()
```
